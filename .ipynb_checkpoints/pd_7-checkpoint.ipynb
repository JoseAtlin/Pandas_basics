```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
knows_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python', na=False).value_counts(normalize=True))
knows_python.rename({True:'Do know', False:'Don\'t know'}, inplace=True)
knows_python.head(10)
```

```python
Num_of_respondents = df['Country'].value_counts()
Num_of_respondents
```

```python
knows_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
knows_python
```

```python
python_stats = pd.concat([Num_of_respondents, knows_python], axis='columns')
python_stats.rename(columns={'Country': 'Num_of_respondents', 'LanguageWorkedWith': 'knows_python'}, inplace=True)
python_stats.head(10)
```

```python
python_stats['percent_python'] = (python_stats['knows_python'] / python_stats['Num_of_respondents']) * 100
python_stats.sort_values('knows_python', ascending=False).head(10)
```

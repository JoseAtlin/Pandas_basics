```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
        'first': ['jose', 'atlin', 'babu'],
        'last': ['jose', 'atlin', 'babu'],
        'email': ['jose@gmail.com', 'atlin@gmail.com', 'babu@gmail.com']
    })

df2 = pd.DataFrame({
        'first': ['raphael', 'andelin'],
        'last': ['raphael', 'andelin'],
        'email': ['raphael@gmail.com', 'andelin@gmail.com']
    })
```

```python
df1['full_name'] = df1['first'] + ' ' + df1['last']
df1
```

```python
df1.drop(columns=['first', 'last'], inplace=True)
df1
```

```python
df1[['first', 'last']] = df1['full_name'].str.split(expand=True)
df1
```

```python
df1.append({'first': 'raphael', 'last': 'raphael'}, ignore_index=True)
```

```python
df1 = df1.append(df2, ignore_index=True, sort=True)
df1
```

```python
df1.drop(index=4)
```

```python
filt = df1['first'] == 'jose'
df1.drop(index=df1[filt].index)
```
